```python
#Bài 18
# Nhập chiều cao n
n = int(input("Nhập chiều cao n: ")) # với hình như yêu cầu của bài tập này thì n=5

print("Hình 1:")
for i in range(n):
    for j in range(n):
        if i == 0 or i == n - 1 or j == 0 or j == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()

print("\nHình 2:")
for i in range(n):
    for j in range(n):
        if i == j or i == 0 or i == n - 1 or j == 0 or j == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()

print("\nHình 3:")
for i in range(n):
    for j in range(n):
        if i + j == n - 1 or i == 0 or j == 0 or j == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
#Nhập một số n có tối đa 2 chữ số. Hãy cho biết cách đọc ra dạng chữ.(vd: n=35 => Ba mươi lăm, n=5 => năm).
n = int(input("Nhập số nguyên n (0 <= n <= 99): "))

don_vi = ["", "một", "hai", "ba", "bốn", "năm", "sáu", "bảy", "tám", "chín"]

if n < 0 or n > 99:
    print("Số không hợp lệ!")
elif n < 10:
    if n == 0:
        print("Không")
    else:
        print(don_vi[n].capitalize())
elif n < 20:
    if n == 10:
        print("Mười")
    elif n == 15:
        print("Mười lăm")
    else:
        print("Mười " + don_vi[n % 10])
else:
    chuc = n // 10
    dv = n % 10

    doc = don_vi[chuc].capitalize() + " mươi"
    if dv == 0:
        print(doc)
    elif dv == 1:
        print(doc + " mốt")
    elif dv == 5:
        print(doc + " lăm")
    else:
        print(doc + " " + don_vi[dv])

```

```python
#Câu 11: Kiểm tra số nguyên tố
n = int(input("Nhập một số nguyên: "))

if n < 2:
    print(n, "không phải là số nguyên tố.")
else:
    for i in range(2, n):
        if n % i == 0:
            print(n, "không phải là số nguyên tố.")
            break
    else:
        # else của for, chỉ chạy khi vòng lặp không bị break
        print(n, "là số nguyên tố.")
```
