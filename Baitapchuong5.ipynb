```python
#cau 1
def CheckDoiXung(s):
  flag=True
  for i in range(len(s)):
    if s[i]!=s[len(s)-i-1]:
      flag=False
      break
  return flag
def main():
  print("Nhập 1 chuỗi:")
  s=input()
  if(CheckDoiXung(s)):
    print("Chuỗi bạn nhập đối xứng")
  else:
    print("Chuỗi bạn nhập không đối xứng")
while True:
  main()
  print("Tiếp không?(c/k):")
  s=input()
  if s=="k":
    break
print("CÁM ƠN ")
```

```python
#câu 4

# Chuỗi ban đầu
my_string = "  Xin Chào Thế Giới Python!  "

# 1. len()
print(f"Độ dài của chuỗi: {len(my_string)}")

# 2. lower()
print(f"Chuỗi sau khi chuyển sang chữ thường: {my_string.lower()}")

# 3. upper()
print(f"Chuỗi sau khi chuyển sang chữ hoa: {my_string.upper()}")

# 4. strip()
stripped_string = my_string.strip()
print(f"Chuỗi sau khi xóa khoảng trắng ở đầu và cuối: '{stripped_string}'")

# 5. split()
sentence = "Python là ngôn ngữ lập trình mạnh mẽ và dễ học."
words = sentence.split(" ")
print(f"Danh sách các từ trong câu: {words}")

# 6. join()
joined_string = "-".join(words)
print(f"Nối các từ bằng dấu gạch ngang: {joined_string}")

# 7. replace()
replaced_string = my_string.replace("Python", "AI")
print(f"Chuỗi sau khi thay thế 'Python' bằng 'AI': {replaced_string}")

# 8. find()
find_index = my_string.find("Thế Giới")
print(f"Chỉ số của 'Thế Giới' trong chuỗi: {find_index}")

# 9. startswith()
starts_with = my_string.startswith("  Xin")
print(f"Chuỗi có bắt đầu bằng '  Xin' không?: {starts_with}")

# 10. endswith()
ends_with = my_string.endswith("!  ")
print(f"Chuỗi có kết thúc bằng '!  ' không?: {ends_with}")
```

```python
#Câu 5
def analyze_string(s):
    upper_count = 0
    lower_count = 0
    digit_count = 0
    special_count = 0
    space_count = 0
    vowel_count = 0
    consonant_count = 0

    vowels = "AEIOUaeiou"
    consonants = "BCDFGHJKLMNPQRSTVWXYZbcdfghjklmnpqrstvwxyz"

    for char in s:
        if 'A' <= char <= 'Z':
            upper_count += 1
            if char in vowels:
                vowel_count += 1
            elif char in consonants:
                consonant_count += 1
        elif 'a' <= char <= 'z':
            lower_count += 1
            if char in vowels:
                vowel_count += 1
            elif char in consonants:
                consonant_count += 1
        elif '0' <= char <= '9':
            digit_count += 1
        elif char == ' ':
            space_count += 1
        else:
            special_count += 1

    print(f"- Số chữ IN HOA: {upper_count}")
    print(f"- Số chữ in thường: {lower_count}")
    print(f"- Số chữ là chữ số: {digit_count}")
    print(f"- Số chữ là ký tự đặc biệt: {special_count}")
    print(f"- Số chữ là khoảng trắng: {space_count}")
    print(f"- Số chữ là Nguyên Âm: {vowel_count}")
    print(f"- Số chữ là Phụ âm: {consonant_count}")

# Nhập chuỗi từ người dùng
input_string = input("Nhập vào 1 chuỗi: ")

# Phân tích chuỗi và in kết quả
analyze_string(input_string)
```

```python
#Câu 6
import re

def NegativeNumberInStrings(s):
  """
  Tìm và in ra các số nguyên âm trong chuỗi.

  Args:
    s: Chuỗi đầu vào.
  """
  # Sử dụng biểu thức chính quy để tìm các số nguyên âm
  negative_numbers = re.findall(r'-\d+', s)

  # In ra các số nguyên âm tìm được
  if negative_numbers:
    print("Các số nguyên âm trong chuỗi là:")
    for number in negative_numbers:
      print(number)
  else:
    print("Không tìm thấy số nguyên âm nào trong chuỗi.")

# Ví dụ sử dụng hàm
input_string = "abc-5xyz-12k9l--p"
NegativeNumberInStrings(input_string)

input_string_2 = "This string has no negative numbers."
NegativeNumberInStrings(input_string_2)
```

```python
#Câu 7
import os

def get_filename_with_extension(filepath):
  """
  Lấy tên file bao gồm cả đuôi mở rộng từ đường dẫn.

  Args:
    filepath: Đường dẫn của file.

  Returns:
    Tên file bao gồm cả đuôi mở rộng.
  """
  return os.path.basename(filepath)

def get_filename_without_extension(filepath):
  """
  Lấy tên file không bao gồm đuôi mở rộng từ đường dẫn.

  Args:
    filepath: Đường dẫn của file.

  Returns:
    Tên file không bao gồm đuôi mở rộng.
  """
  filename_with_ext = os.path.basename(filepath)
  filename_without_ext, _ = os.path.splitext(filename_with_ext)
  return filename_without_ext

# Ví dụ sử dụng
filepath = "d:\\music\\muabui.mp3"

filename_with_ext = get_filename_with_extension(filepath)
print(f"Tên file có đuôi mở rộng: {filename_with_ext}")

filename_without_ext = get_filename_without_extension(filepath)
print(f"Tên file không có đuôi mở rộng: {filename_without_ext}")

filepath_2 = "/usr/local/songs/amazing_grace.wav"
filename_with_ext_2 = get_filename_with_extension(filepath_2)
print(f"Tên file có đuôi mở rộng: {filename_with_ext_2}")

filename_without_ext_2 = get_filename_without_extension(filepath_2)
print(f"Tên file không có đuôi mở rộng: {filename_without_ext_2}")
```

```python
#câu 8
def optimize_string(s):
  """
  Tối ưu chuỗi danh từ: xóa khoảng trắng thừa, cách nhau 1 khoảng trắng, viết hoa chữ cái đầu mỗi từ.

  Args:
    s: Chuỗi đầu vào.

  Returns:
    Chuỗi đã được tối ưu.
  """
  # Xóa khoảng trắng ở đầu và cuối chuỗi và chuyển về chữ thường
  s = s.strip().lower()

  # Tách chuỗi thành các từ
  words = s.split()

  # Viết hoa chữ cái đầu của mỗi từ và nối lại bằng một khoảng trắng
  optimized_words = [word.capitalize() for word in words]
  optimized_string = " ".join(optimized_words)

  return optimized_string

# Ví dụ sử dụng hàm
input_string = " TRần duY thAnH "
output_string = optimize_string(input_string)
print(f'Input "{input_string}"')
print(f'Output "{output_string}"')

input_string_2 = "   nguyen   van   a   "
output_string_2 = optimize_string(input_string_2)
print(f'Input "{input_string_2}"')
print(f'Output "{output_string_2}"')
```
