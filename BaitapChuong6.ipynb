```python
#câu 6
def get_increasing_list():
    """Prompts the user to enter numbers in increasing order and validates input."""
    input_list = []
    while True:
        try:
            num_str = input("Enter a number (or 'done' to finish): ")
            if num_str.lower() == 'done':
                break
            num = float(num_str)
            if not input_list or num >= input_list[-1]:
                input_list.append(num)
            else:
                print("Error: Please enter a number greater than or equal to the previous one.")
        except ValueError:
            print("Invalid input. Please enter a valid number or 'done'.")
    return input_list

print("Enter numbers in increasing order. Type 'done' when finished.")
sorted_list = get_increasing_list()
print("The entered sorted list is:")
print(sorted_list)
```

```python
#câu 7
import random

def generate_unique_random_list(n, lower_bound, upper_bound):
  """Generates a list of n unique random numbers within a specified range."""
  if n > (upper_bound - lower_bound + 1):
    return "Error: Cannot generate N unique numbers in the given range."

  unique_numbers = set()
  while len(unique_numbers) < n:
    unique_numbers.add(random.randint(lower_bound, upper_bound))

  return list(unique_numbers)

# Get input from the user
n = int(input("Enter the number of unique random elements (N): "))
lower_bound = int(input("Enter the lower bound for the random numbers: "))
upper_bound = int(input("Enter the upper bound for the random numbers: "))

# Generate and print the list
random_list = generate_unique_random_list(n, lower_bound, upper_bound)
print("Generated list of unique random numbers:")
print(random_list)
```

```python
#câu 8
def sort_list_descending():
  """Prompts the user to enter real numbers and sorts them in descending order."""
  input_list = []
  print("Enter real numbers one by one (type 'done' when finished):")
  while True:
    num_str = input()
    if num_str.lower() == 'done':
      break
    try:
      input_list.append(float(num_str))
    except ValueError:
      print("Invalid input. Please enter a valid real number or 'done'.")

  input_list.sort(reverse=True)
  return input_list

sorted_list_descending = sort_list_descending()
print("The list sorted in descending order is:")
print(sorted_list_descending)
```
