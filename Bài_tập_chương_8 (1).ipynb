```python
#Bài 2
from tkinter import *
from math import sqrt

def giaiAction():
    a = float(stringHSA.get())
    b = float(stringHSB.get())
    c = float(stringHSC.get())

    if a == 0:  # bx + c = 0
        if b == 0 and c == 0:
            stringKQ.set("Vô số nghiệm")
        elif b == 0 and c != 0:
            stringKQ.set("Vô nghiệm")
        else:
            x = -c / b
            stringKQ.set("x = " + str(x))
    else:
        delta = b**2 - 4*a*c
        if delta < 0:
            stringKQ.set("Vô nghiệm")
        elif delta == 0:
            x = -b / (2*a)
            stringKQ.set("Nghiệm kép x1 = x2 = " + str(x))
        else:
            x1 = (-b - sqrt(delta)) / (2*a)
            x2 = (-b + sqrt(delta)) / (2*a)
            stringKQ.set("x1 = " + str(x1) + "; x2 = " + str(x2))

def tiepAction():
    stringHSA.set("")
    stringHSB.set("")
    stringHSC.set("")
    stringKQ.set("")

root = Tk()
stringHSA = StringVar()
stringHSB = StringVar()
stringHSC = StringVar()
stringKQ = StringVar()

root.title("Giải Phương Trình Bậc 2")
root.minsize(height=180, width=300)

Label(root, text="Phương Trình Bậc 2", fg="red",
      font=("tahoma", 16)).grid(row=0, column=0, columnspan=2)

Label(root, text="Hệ số a:").grid(row=1, column=0)
Entry(root, width=30, textvariable=stringHSA).grid(row=1, column=1)

Label(root, text="Hệ số b:").grid(row=2, column=0)
Entry(root, width=30, textvariable=stringHSB).grid(row=2, column=1)

Label(root, text="Hệ số c:").grid(row=3, column=0)
Entry(root, width=30, textvariable=stringHSC).grid(row=3, column=1)

frameButton = Frame(root)
Button(frameButton, text="Giải", command=giaiAction).pack(side=LEFT)
Button(frameButton, text="Tiếp", command=tiepAction).pack(side=LEFT)
Button(frameButton, text="Thoát", command=root.quit).pack(side=LEFT)
frameButton.grid(row=4, columnspan=2)

Label(root, text="Kết quả:").grid(row=5, column=0)
Entry(root, width=30, textvariable=stringKQ).grid(row=5, column=1)

root.mainloop()
```

```python
#Bài 3
from tkinter import *

def congAction():
    a = float(stringA.get())
    b = float(stringB.get())
    stringKQ.set(a + b)

def truAction():
    a = float(stringA.get())
    b = float(stringB.get())
    stringKQ.set(a - b)

def nhanAction():
    a = float(stringA.get())
    b = float(stringB.get())
    stringKQ.set(a * b)

def chiaAction():
    a = float(stringA.get())
    b = float(stringB.get())
    if b == 0:
        stringKQ.set("Lỗi: chia 0")
    else:
        stringKQ.set(a / b)

root = Tk()
stringA = StringVar()
stringB = StringVar()
stringKQ = StringVar()

root.minsize(height=150, width=250)
root.title("Cộng Trừ Nhân Chia")

Label(root, text="Cộng Trừ Nhân Chia", fg="blue",
      font=("tahoma", 16)).grid(row=0, columnspan=3)

frameButton = Frame(root)
Button(frameButton, text="Cộng", command=congAction).pack(side=TOP, fill=X)
Button(frameButton, text="Trừ", command=truAction).pack(side=TOP, fill=X)
Button(frameButton, text="Nhân", command=nhanAction).pack(side=TOP, fill=X)
Button(frameButton, text="Chia", command=chiaAction).pack(side=TOP, fill=X)
frameButton.grid(row=1, column=0, rowspan=4)

Label(root, text="Số a:").grid(row=1, column=1)
Entry(root, width=15, textvariable=stringA).grid(row=1, column=2)

Label(root, text="Số b:").grid(row=2, column=1)
Entry(root, width=15, textvariable=stringB).grid(row=2, column=2)

Label(root, text="Kết quả:").grid(row=3, column=1)
Entry(root, width=15, textvariable=stringKQ).grid(row=3, column=2)

Button(root, text="Thoát", command=root.quit).grid(row=4, column=2)

root.mainloop()
```

```python
#Bài 4
from tkinter import *

def button_click(x):
    current = stringKQ.get()
    stringKQ.set(current + x)

def button_clear():
    stringKQ.set("")

def button_equal():
    try:
        result = eval(stringKQ.get())
        stringKQ.set(result)
    except:
        stringKQ.set("Lỗi")

root = Tk()
root.title("Calculator")

stringKQ = StringVar()

Entry(root, textvariable=stringKQ, width=20, font=("tahoma", 16), justify="right").grid(row=0, column=0, columnspan=4)

buttons = [
    ("7",1,0),("8",1,1),("9",1,2),("/",1,3),
    ("4",2,0),("5",2,1),("6",2,2),("*",2,3),
    ("1",3,0),("2",3,1),("3",3,2),("-",3,3),
    ("0",4,0),(".",4,1),("=",4,2),("+",4,3),
]

for (txt, r, c) in buttons:
    if txt == "=":
        Button(root, text=txt, width=5, height=2, command=button_equal).grid(row=r, column=c)
    else:
        Button(root, text=txt, width=5, height=2, command=lambda x=txt: button_click(x)).grid(row=r, column=c)

Button(root, text="Clr", width=22, height=2, command=button_clear).grid(row=5, column=0, columnspan=4)

root.mainloop()
```

```python
#Bài 5
from tkinter import *
from tkinter import messagebox

def login():
    user = stringUser.get()
    pw = stringPass.get()
    if user == "admin" and pw == "123":
        messagebox.showinfo("Thông báo", "Đăng nhập thành công!")
    else:
        messagebox.showerror("Lỗi", "Sai username hoặc password!")

def cancel():
    root.quit()

root = Tk()
root.title("Login Screen")

stringUser = StringVar()
stringPass = StringVar()

Label(root, text="User Name:").grid(row=0, column=0, padx=10, pady=5)
Entry(root, textvariable=stringUser).grid(row=0, column=1, pady=5)

Label(root, text="Password:").grid(row=1, column=0, padx=10, pady=5)
Entry(root, textvariable=stringPass, show="*").grid(row=1, column=1, pady=5)

Button(root, text="Login", width=10, command=login).grid(row=2, column=0, pady=10)
Button(root, text="Cancel", width=10, command=cancel).grid(row=2, column=1)

root.mainloop()
```

```python
#Bài 7
from tkinter import *

def convertAction():
    year = int(stringDL.get())
    can = ["Giáp","Ất","Bính","Đinh","Mậu","Kỷ","Canh","Tân","Nhâm","Quý"]
    chi = ["Tý","Sửu","Dần","Mão","Thìn","Tỵ","Ngọ","Mùi","Thân","Dậu","Tuất","Hợi"]

    nam_can = can[(year + 6) % 10]
    nam_chi = chi[(year + 8) % 12]

    stringAL.set(nam_can + " " + nam_chi)

root = Tk()
root.title("Chuyển Dương Lịch sang Âm Lịch")
root.configure(bg="yellow")

stringDL = StringVar()
stringAL = StringVar()

Label(root, text="Nhập năm dương:", bg="yellow").grid(row=0, column=0)
Entry(root, textvariable=stringDL).grid(row=0, column=1)

Button(root, text="Chuyển", command=convertAction).grid(row=1, column=1)

Label(root, text="Năm âm:", bg="yellow").grid(row=2, column=0)
Entry(root, textvariable=stringAL).grid(row=2, column=1)

root.mainloop()
```

```python
#Bài 8
from tkinter import *

def convertFtoC():
    f = float(stringF.get())
    c = (f - 32) * 5 / 9
    stringC.set(round(c, 2))

root = Tk()
root.title("Đổi độ F sang độ C")
root.configure(bg="yellow")

stringF = StringVar()
stringC = StringVar()

Label(root, text="Nhập độ F:", bg="yellow").grid(row=0, column=0)
Entry(root, textvariable=stringF).grid(row=0, column=1)

Button(root, text="Chuyển", command=convertFtoC).grid(row=1, column=1)

Label(root, text="Độ C:", bg="yellow").grid(row=2, column=0)
Entry(root, textvariable=stringC).grid(row=2, column=1)

root.mainloop()
```

```python
#Bài 9
from tkinter import *

def tinhBMI():
    cao = float(stringCao.get())
    nang = float(stringNang.get())
    bmi = nang / (cao ** 2)
    stringBMI.set(round(bmi, 2))

    if bmi < 18.5:
        stringTT.set("Gầy")
    elif bmi < 25:
        stringTT.set("Bình thường")
    elif bmi < 30:
        stringTT.set("Hơi Béo")
    else:
        stringTT.set("Béo phì")

def thoat():
    root.quit()

root = Tk()
root.title("Tính BMI")
root.configure(bg="yellow")

stringCao = StringVar()
stringNang = StringVar()
stringBMI = StringVar()
stringTT = StringVar()

Label(root, text="Nhập chiều cao (m):", bg="yellow").grid(row=0, column=0)
Entry(root, textvariable=stringCao).grid(row=0, column=1)

Label(root, text="Nhập cân nặng (kg):", bg="yellow").grid(row=1, column=0)
Entry(root, textvariable=stringNang).grid(row=1, column=1)

Button(root, text="Tính BMI", command=tinhBMI).grid(row=2, column=1)

Label(root, text="BMI của bạn:", bg="yellow").grid(row=3, column=0)
Entry(root, textvariable=stringBMI).grid(row=3, column=1)

Label(root, text="Tình trạng:", bg="yellow").grid(row=4, column=0)
Entry(root, textvariable=stringTT).grid(row=4, column=1)

Button(root, text="Thoát", command=thoat).grid(row=5, column=1)

root.mainloop()
```
