```python
#Bài 9
class Category:
    def __init__(self, code, name):
        self.code = code
        self.name = name

    def __str__(self):
        return f"Category Code: {self.code}, Name: {self.name}"

class Product:
    def __init__(self, code, name, price, category_code):
        self.code = code
        self.name = name
        self.price = price
        self.category_code = category_code

    def __str__(self):
        return f"Product Code: {self.code}, Name: {self.name}, Price: {self.price}, Category Code: {self.category_code}"

print("Category and Product classes defined successfully.")
```

```python
import os

# 1. Định nghĩa hàm save_categories_to_file
def save_categories_to_file(file_path, categories):
    with open(file_path, 'w', encoding='utf-8') as file:
        for category in categories:
            file.write(f"{category.code};{category.name}\n")
    print(f"Saved {len(categories)} categories to {file_path}")

# 2. Định nghĩa hàm load_categories_from_file
def load_categories_from_file(file_path):
    loaded_categories = []
    if not os.path.exists(file_path):
        print(f"File '{file_path}' not found. Returning empty list.")
        return loaded_categories
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                line = line.strip()
                if line:
                    parts = line.split(';')
                    if len(parts) == 2:
                        code, name = parts[0], parts[1]
                        loaded_categories.append(Category(code, name))
                    else:
                        print(f"Warning: Skipping malformed line in {file_path}: {line}")
        print(f"Loaded {len(loaded_categories)} categories from {file_path}")
    except Exception as e:
        print(f"Error reading file {file_path}: {e}")
    return loaded_categories

# 3. Tạo một danh sách mẫu gồm 2-3 đối tượng Category
sample_categories = [
    Category("DM01", "Điện Tử"),
    Category("DM02", "Gia Dụng"),
    Category("DM03", "Thời Trang")
]
print("Sample categories created:")
for cat in sample_categories:
    print(cat)
print()

# 4. Gọi hàm save_categories_to_file để lưu danh sách mẫu này vào một file văn bản có tên 'categories.txt'
save_categories_to_file('categories.txt', sample_categories)
print()

# 5. Gọi hàm load_categories_from_file để đọc danh sách danh mục từ file 'categories.txt'
loaded_categories_from_file = load_categories_from_file('categories.txt')
print()

# 6. In ra danh sách các đối tượng Category đã đọc được để kiểm tra tính đúng đắn.
print("Categories loaded from file:")
if loaded_categories_from_file:
    for cat in loaded_categories_from_file:
        print(cat)
else:
    print("No categories were loaded.")
```

```python
import os

# Assuming Category and Product classes are already defined
# (from cell 1cc7c16b)
# Assuming save_categories_to_file, load_categories_from_file, save_products_to_file,
# and load_products_from_file are already defined (from cell 088a164d and 8d8fc3f6)

# --- Product Management Functions ---

def display_products(products):
    """In ra danh sách các đối tượng Product."""
    if not products:
        print("No products to display.")
        return
    print("--- Current Products ---")
    for product in products:
        print(product)
    print("------------------------")

def add_product(products, categories, code, name, price, category_code):
    """Thêm một sản phẩm mới vào danh sách và lưu vào file."""
    if any(p.code == code for p in products):
        print(f"Error: Product with code '{code}' already exists.")
        return False

    if not any(c.code == category_code for c in categories):
        print(f"Error: Category code '{category_code}' does not exist. Cannot add product.")
        return False

    try:
        price = float(price)
    except ValueError:
        print(f"Error: Invalid price '{price}'. Price must be a number.")
        return False

    new_product = Product(code, name, price, category_code)
    products.append(new_product)
    save_products_to_file('products.txt', products)
    print(f"Product '{name}' ({code}) added successfully.")
    return True

def update_product(products, categories, code, new_name=None, new_price=None, new_category_code=None):
    """Cập nhật thông tin của một sản phẩm và lưu vào file."""
    found = False
    for product in products:
        if product.code == code:
            found = True
            if new_category_code:
                if not any(c.code == new_category_code for c in categories):
                    print(f"Error: Category code '{new_category_code}' does not exist. Cannot update product.")
                    return False
                product.category_code = new_category_code
            if new_name is not None:
                product.name = new_name
            if new_price is not None:
                try:
                    product.price = float(new_price)
                except ValueError:
                    print(f"Error: Invalid price '{new_price}'. Price must be a number.")
                    return False
            break

    if found:
        save_products_to_file('products.txt', products)
        print(f"Product with code '{code}' updated successfully.")
        return True
    else:
        print(f"Error: Product with code '{code}' not found.")
        return False

def delete_product(products, code):
    """Xóa một sản phẩm khỏi danh sách và lưu vào file."""
    initial_len = len(products)
    products[:] = [product for product in products if product.code != code]

    if len(products) < initial_len:
        save_products_to_file('products.txt', products)
        print(f"Product with code '{code}' deleted successfully.")
        return True
    else:
        print(f"Error: Product with code '{code}' not found.")
        return False

def find_products(products, keyword):
    """Tìm kiếm các sản phẩm có code, name, price (as string), hoặc category_code chứa keyword (không phân biệt chữ hoa, chữ thường)."""
    keyword_lower = str(keyword).lower()
    found_products = [
        prod for prod in products
        if keyword_lower in prod.code.lower() or \
           keyword_lower in prod.name.lower() or \
           keyword_lower in str(prod.price).lower() or \
           keyword_lower in prod.category_code.lower()
    ]
    return found_products

def sort_products(products, key, reverse=False):
    """Sắp xếp danh sách sản phẩm dựa trên thuộc tính key ('code', 'name', 'price', hoặc 'category_code')."""
    if key not in ['code', 'name', 'price', 'category_code']:
        print(f"Error: Invalid key '{key}' for sorting. Choose 'code', 'name', 'price', or 'category_code'.")
        return list(products) # Return a copy to avoid modifying original if key is invalid

    if key == 'price':
        sorted_list = sorted(products, key=lambda prod: prod.price, reverse=reverse)
    else:
        sorted_list = sorted(products, key=lambda prod: getattr(prod, key).lower(), reverse=reverse)

    print(f"Products sorted by '{key}' (reverse={reverse}).")
    return sorted_list

# --- Initialization and Demonstrations ---

# 1. Tải danh sách danh mục và sản phẩm hiện có
categories = load_categories_from_file('categories.txt')
products = load_products_from_file('products.txt')

print("\n--- Initial Data ---")
display_categories(categories) # Re-use display_categories from previous step
display_products(products)

# 2. Thêm mới sản phẩm
print("\n--- Adding New Products ---")
add_product(products, categories, "SP004", "Bàn Phím Cơ", 99.99, "DM01") # Valid
display_products(products)

add_product(products, categories, "SP001", "Test Duplicate", 1.0, "DM01") # Existing code
add_product(products, categories, "SP005", "Ghế Văn Phòng", 150.00, "DM99") # Invalid category
add_product(products, categories, "SP006", "Loa Bluetooth", "invalid_price", "DM01") # Invalid price
display_products(products)

# 3. Sửa sản phẩm
print("\n--- Updating Products ---")
update_product(products, categories, "SP004", new_name="Bàn Phím Gaming", new_price=120.00, new_category_code="DM02") # Valid update
display_products(products)

update_product(products, categories, "SP004", new_category_code="DM99") # Invalid category code for update
update_product(products, categories, "SP007", new_name="Non Existent", new_price=10.0) # Non-existent product
update_product(products, categories, "SP004", new_price="invalid_price") # Invalid price for update
display_products(products)

# 4. Tìm kiếm sản phẩm
print("\n--- Searching Products ---")
found_by_name_or_code = find_products(products, "SP001")
print(f"Products found with keyword 'SP001':")
display_products(found_by_name_or_code)

found_by_partial_name = find_products(products, "máy")
print(f"Products found with keyword 'máy':")
display_products(found_by_partial_name)

found_by_price = find_products(products, 120)
print(f"Products found with keyword '120':")
display_products(found_by_price)

found_by_category = find_products(products, "dm02")
print(f"Products found with keyword 'dm02':")
display_products(found_by_category)

# 5. Sắp xếp sản phẩm
print("\n--- Sorting Products ---")
sorted_by_price = sort_products(products, 'price', reverse=True)
print("Sorted by Price (Descending):")
display_products(sorted_by_price)

sorted_by_name = sort_products(products, 'name')
print("Sorted by Name (Ascending):")
display_products(sorted_by_name)

sorted_by_category_code = sort_products(products, 'category_code')
print("Sorted by Category Code (Ascending):")
display_products(sorted_by_category_code)

# 6. Xóa sản phẩm
print("\n--- Deleting Products ---")
delete_product(products, "SP004") # Valid delete
display_products(products)

delete_product(products, "SP007") # Non-existent product
display_products(products)

print("\n--- Final Products ---")
display_products(products)
```

```python
#Bài 10
import json

# ===================== Dữ liệu =====================
data_file = "students.json"

# Danh sách lớp và sinh viên
classes = {}

# ===================== Hàm lưu và đọc JSON =====================
def save_to_file():
    with open(data_file, "w") as f:
        json.dump(classes, f, indent=4)
    print("Đã lưu dữ liệu vào file JSON.")

def load_from_file():
    global classes
    try:
        with open(data_file, "r") as f:
            classes = json.load(f)
        print("Đã đọc dữ liệu từ file JSON.")
    except FileNotFoundError:
        print("File JSON chưa tồn tại, sẽ tạo mới khi lưu.")
        classes = {}

# ===================== Hàm quản lý lớp =====================
def add_class():
    class_id = input("Nhập mã lớp: ")
    class_name = input("Nhập tên lớp: ")
    if class_id in classes:
        print("Lớp đã tồn tại!")
        return
    classes[class_id] = {"name": class_name, "students": []}
    print(f"Đã thêm lớp {class_name}.")

def show_classes():
    if not classes:
        print("Chưa có lớp nào.")
        return
    print("Danh sách lớp:")
    for cid, c in classes.items():
        print(f"- {cid}: {c['name']} (Số sinh viên: {len(c['students'])})")

# ===================== Hàm quản lý sinh viên =====================
def add_student():
    class_id = input("Nhập mã lớp sinh viên: ")
    if class_id not in classes:
        print("Lớp không tồn tại!")
        return
    student_id = input("Nhập mã sinh viên: ")
    student_name = input("Nhập tên sinh viên: ")
    birth_year = input("Nhập năm sinh: ")
    # Kiểm tra trùng mã sinh viên trong lớp
    for s in classes[class_id]["students"]:
        if s["id"] == student_id:
            print("Mã sinh viên đã tồn tại!")
            return
    classes[class_id]["students"].append({
        "id": student_id,
        "name": student_name,
        "birth_year": birth_year
    })
    print(f"Đã thêm sinh viên {student_name} vào lớp {classes[class_id]['name']}.")

def show_students():
    class_id = input("Nhập mã lớp muốn xem sinh viên: ")
    if class_id not in classes:
        print("Lớp không tồn tại!")
        return
    students = classes[class_id]["students"]
    if not students:
        print("Chưa có sinh viên nào trong lớp.")
        return
    print(f"Sinh viên trong lớp {classes[class_id]['name']}:")
    for s in students:
        print(f"- {s['id']}: {s['name']} ({s['birth_year']})")

def edit_student():
    class_id = input("Nhập mã lớp: ")
    if class_id not in classes:
        print("Lớp không tồn tại!")
        return
    student_id = input("Nhập mã sinh viên muốn sửa: ")
    for s in classes[class_id]["students"]:
        if s["id"] == student_id:
            s["name"] = input(f"Nhập tên mới ({s['name']}): ") or s["name"]
            s["birth_year"] = input(f"Nhập năm sinh mới ({s['birth_year']}): ") or s["birth_year"]
            print("Đã cập nhật sinh viên.")
            return
    print("Không tìm thấy sinh viên.")

def delete_student():
    class_id = input("Nhập mã lớp: ")
    if class_id not in classes:
        print("Lớp không tồn tại!")
        return
    student_id = input("Nhập mã sinh viên muốn xóa: ")
    for s in classes[class_id]["students"]:
        if s["id"] == student_id:
            classes[class_id]["students"].remove(s)
            print("Đã xóa sinh viên.")
            return
    print("Không tìm thấy sinh viên.")

def search_student():
    keyword = input("Nhập tên hoặc mã sinh viên muốn tìm: ").lower()
    found = False
    for cid, c in classes.items():
        for s in c["students"]:
            if keyword in s["id"].lower() or keyword in s["name"].lower():
                print(f"Lớp {c['name']} - {s['id']}: {s['name']} ({s['birth_year']})")
                found = True
    if not found:
        print("Không tìm thấy sinh viên.")

def sort_students():
    class_id = input("Nhập mã lớp muốn sắp xếp: ")
    if class_id not in classes:
        print("Lớp không tồn tại!")
        return
    classes[class_id]["students"].sort(key=lambda s: s["name"])
    print(f"Đã sắp xếp sinh viên trong lớp {classes[class_id]['name']} theo tên.")

# ===================== Menu chính =====================
def menu():
    load_from_file()
    while True:
        print("\n--- QUẢN LÝ SINH VIÊN ---")
        print("1. Thêm lớp")
        print("2. Xem lớp")
        print("3. Thêm sinh viên")
        print("4. Xem sinh viên")
        print("5. Sửa sinh viên")
        print("6. Xóa sinh viên")
        print("7. Tìm kiếm sinh viên")
        print("8. Sắp xếp sinh viên theo tên")
        print("9. Lưu dữ liệu")
        print("0. Thoát")
        choice = input("Chọn chức năng: ")
        if choice == "1":
            add_class()
        elif choice == "2":
            show_classes()
        elif choice == "3":
            add_student()
        elif choice == "4":
            show_students()
        elif choice == "5":
            edit_student()
        elif choice == "6":
            delete_student()
        elif choice == "7":
            search_student()
        elif choice == "8":
            sort_students()
        elif choice == "9":
            save_to_file()
        elif choice == "0":
            print("Thoát chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ!")

# ===================== Chạy chương trình =====================
if __name__ == "__main__":
    menu()
```

```python
#Bài 11
import pandas as pd
import os

# File Excel
excel_file = "NhanVien.xlsx"

# ===================== Hàm lưu nhân viên =====================
def save_employees(employees):
    df = pd.DataFrame(employees)
    df.to_excel(excel_file, index=False)
    print(f"Đã lưu {len(employees)} nhân viên vào file {excel_file}.")

# ===================== Hàm đọc nhân viên =====================
def load_employees():
    if os.path.exists(excel_file):
        df = pd.read_excel(excel_file)
        employees = df.to_dict(orient="records")
        return employees
    else:
        print("File Excel chưa tồn tại.")
        return []

# ===================== Hàm hiển thị nhân viên =====================
def show_employees(employees):
    if not employees:
        print("Chưa có nhân viên nào.")
        return
    print("Danh sách nhân viên:")
    for e in employees:
        print(f"{e['Mã']} - {e['Tên']} - {e['Tuổi']}")

# ===================== Hàm thêm nhân viên =====================
def add_employee(employees):
    emp_id = input("Nhập mã nhân viên: ")
    name = input("Nhập tên nhân viên: ")
    age = int(input("Nhập tuổi nhân viên: "))
    employees.append({"Mã": emp_id, "Tên": name, "Tuổi": age})
    print("Đã thêm nhân viên.")

# ===================== Hàm sắp xếp nhân viên =====================
def sort_by_age(employees):
    employees.sort(key=lambda x: x["Tuổi"])
    print("Đã sắp xếp nhân viên theo tuổi tăng dần.")

# ===================== Menu chính =====================
def menu():
    employees = load_employees()
    while True:
        print("\n--- QUẢN LÝ NHÂN VIÊN ---")
        print("1. Xem danh sách nhân viên")
        print("2. Thêm nhân viên")
        print("3. Sắp xếp nhân viên theo tuổi")
        print("4. Lưu dữ liệu vào Excel")
        print("0. Thoát")
        choice = input("Chọn chức năng: ")
        if choice == "1":
            show_employees(employees)
        elif choice == "2":
            add_employee(employees)
        elif choice == "3":
            sort_by_age(employees)
        elif choice == "4":
            save_employees(employees)
        elif choice == "0":
            break
        else:
            print("Chọn không hợp lệ!")

if __name__ == "__main__":
    menu()
```
